```python
from langchain.document_loaders import PyPDFLoader
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
loader = PyPDFLoader(r'D:\College\LLMOPS\document_portal\Books\Basic-Electrical-Engineering.pdf')
documents = loader.load()
```

```python
documents
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)
```

```python
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
docs[0].metadata
```

```python
docs[0].metadata['source'].split('\\')[-1]
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
embeddings_model = GoogleGenerativeAIEmbeddings(
    model = 'models/embedding-001',
    max_retries=5
)
```

```python
embeddings_model.embed_query("What is the average salary of a Data science(Intern) in India?")
```

```python
from langchain.vectorstores import FAISS
```

```python
vectorestore = FAISS.from_documents(docs,embeddings_model)
```

```python
relevant_doc=vectorestore.similarity_search("What is the Switching characteristics of thyristors?",k= 5)
```

```python
relevant_doc
```

```python
retriver = vectorestore.as_retriever(search_type="similarity", search_kwargs={"k": 10})
```

```python
retriver.invoke("What is the Switching characteristics of thyristors?")
```

```python
propmt_template = """
You are a helpful assistant. You will be provided with some context and you need to answer the question based on that context.
If the context does not contain sufficient information, respond with:
"I don't have enough information about this."
Context: {context}
Question: {question}

"""
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
    template=propmt_template,
    input_variables=["context", "question"]
)
```

```python
prompt
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser = StrOutputParser()
```

```python
def formate_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
rag_chain = (
    {"context": retriver | formate_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | parser
)
```

```python
rag_chain.invoke("What is OHM’S LAW ")
```
